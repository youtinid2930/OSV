```python


trainDr = "./data/Train"


X_train, y_train = loadData(trainDr)

print(f"Nombre d'image chargees : {len(X_train)}")

print(f"Forme du tableau X : {X_train.shape}")

print(f"Labels uniques : {set(y_train)}")



for i in range(5):
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label : { 'Forged' if y_train[i] == 1 else 'Original'} ")
    plt.show()
```
